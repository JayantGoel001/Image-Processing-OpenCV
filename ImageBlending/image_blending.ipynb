## Image Blending

```python
import cv2
from matplotlib import pyplot as plot
import numpy as nm
```

```python
img_1 =cv2.imread("mango.jpg")[:,:,::-1]
img_2=cv2.imread("apple.jpg")[:,:,::-1]
```

```python
img_2=cv2.resize(img_2,(img_1.shape[1],img_1.shape[0]))
disp_img=nm.hstack((img_1,img_2))
plot.imshow(disp_img)
plot.title("Original Images")
```

```python
blended_img=cv2.addWeighted(img_1,0.45,img_2,0.55,0)
```

```python
plot.imshow(blended_img)
plot.title("Blended Images")
```
