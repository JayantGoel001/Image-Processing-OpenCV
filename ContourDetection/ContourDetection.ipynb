# Import Modules

```python
import cv2
from google.colab.patches import cv2_imshow
```

# Load the Image

```python
# The image is loaded using cv2.imread
# Syntax: cv2.imread(path, flag)
# Flag specifies how the image should be read
# We set it to 1 to indicate that we want to load the image in color

image = cv2.imread('demo1.png',1)
cv2_imshow(image)
```

## Preprocessing the Image

```python
# The image is first converted to a gray scale image
# This is because thresholding can only be applied to gray-scale images

gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
```

```python
# Simple Thresholding is applied to the image. 
# Applying Thresholding would give us a binary image, 
# using which we could expect a better accuracy in detecting contours.
# syntax: cv2.threshold(source_image, thresh_value, max_value, threshold_type, destination=None)
# thresh_value can be adjusted according to the image

_, thresh = cv2.threshold(gray_image,205,255,0)
cv2_imshow(thresh)
```

# Contour Detection

```python
# Contours are detected using cv2.findContours()
# Syntax: cv2.findContours(image, mode, method, contours=None, hierarchy=None, offset=None)

contours, heirarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
```

# Display Contours

```python
# contours are displayed on the original image by using cv2.drawContours()
# Syntax: cv2.drawContours(image, contours, contourIndex, color, thickness=None,)

# to display all contours in the image, we set index to -1
index = -1

# color of contours is set to blue (BGR)
color = (255,0,0)

# thickness is set to 2
thickness = 2

cv2.drawContours(image, contours, index, color, thickness)
cv2_imshow(image)
```
