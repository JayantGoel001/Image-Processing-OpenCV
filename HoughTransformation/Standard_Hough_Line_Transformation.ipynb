```python
# import the modules
import math
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
# read original image
img = cv.imread("./chess.jpg")

# convert to RGB
src = cv.cvtColor(img, cv.COLOR_BGR2RGB)

# Apply gray filter
gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
```

```python
# show the original image
plt.imshow(src)
plt.title("Original Image")
plt.show()
```

```python
# Edge detection
edges = cv.Canny(gray, 50, 150, None, apertureSize = 3)

# Standard Hough Line Transform
lines = cv.HoughLines(edges, 1, math.pi / 180, 100)

# Draw the lines
for line in lines:
    rho,theta = line[0]
    a = np.cos(theta)
    b = np.sin(theta)
    x0 = a*rho
    y0 = b*rho
    x1 = int(x0 + 1000*(-b))
    y1 = int(y0 + 1000*(a))
    x2 = int(x0 - 1000*(-b))
    y2 = int(y0 - 1000*(a))
    cv.line(src,(x1,y1),(x2,y2),(0,255,0),2)
```

```python
# Display final image
plt.imshow(src)
plt.title("Standard Hough Line Transform")
plt.show()
```
