```python
#import the modules
import cv2
from matplotlib import pyplot as plot
import numpy as np
```

```python
# read original image
img = cv2.imread("C:/Users/HY/Downloads/chess.jpg")

#convert to RGB
src = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

#Apply gray filter
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
# show the original image
plot.imshow(src)
plot.title("Original Image")
plot.show()
```

```python
# Edge detection
edges = cv2.Canny(gray, 50, 150, None, apertureSize = 3)
minLineLength = 100
maxLineGap = 10

# Probabilistic Line Transform
lines = cv2.HoughLinesP(edges, 1, np.pi/180, 100, None, minLineLength, maxLineGap)

# Draw the lines
for line in lines:
    x1,y1,x2,y2 = line[0]
    cv2.line(src,(x1,y1),(x2,y2),(0,255,0),2)
```

```python
# Display final image
plot.imshow(src)
plot.title("Probabilistic Houghlines.jpg")
plot.show()
```
