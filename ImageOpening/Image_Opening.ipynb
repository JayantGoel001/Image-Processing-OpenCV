#Import Modules

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

#Load Image 

```python
#image is loaded using cv2.imread() method,here flag is 0 ,specifies to load image in GRAYSCALE mode.
'''
Syntax:
  cv2.imread(path,flag)
Parameters:
  path: string representing the path of the image to be read.
  flag: specifies the way in which image should be read.
'''
img=cv2.imread("input.jpg",0)
cv2_imshow(img)
```

#Apply Opening Operation




```python
# To perform Opening operation,cv2.morphologyEx() method is used.
'''
Syntax:
  morphologyEx (src, dst, op, kernel, anchor, iterations, borderType, borderValue)
Parameters:
  src: input image.
  dst: output image ; optional parameter.
  op: Type of morphological operation.(open/close)
  kernel: Structuring element used for Closing.
  anchor: Anchor position inside the structuring element. The default value is [-1, -1} signifying position as the center of the structuring element.
  iterations: Number of times Closing is applied.
  borderType: Type of border ( BORDER_CONSTANT, BORDER_REPLICATE etc.)
  borderValue: Border value
'''
kernel=np.ones((5,5),np.uint8)
Opened=cv2.morphologyEx(img,cv2.MORPH_OPEN,kernel,iterations=1)
```

#Display the Opened image

```python
cv2_imshow(Opened)
```
