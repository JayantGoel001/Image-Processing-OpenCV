#Import Modules

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

#Load Image 

```python
#image is loaded using cv2.imread() method,here flag is 1 ,specifies to load a color image.
'''
Syntax:
  cv2.imread(path,flag)
Parameters:
  path: string representing the path of the image to be read.
  flag: specifies the way in which image should be read.
'''
img=cv2.imread("input.jpg",1)
cv2_imshow(img)
```

#Apply Rotation operation


```python
# To perform Rotation,cv2.rotate() method is used.
'''
Syntax:
  cv2.cv.rotate( src, rotateCode[, dst] )
Parameters:
  src: Input image that is to be rotated.
  rotateCode: An enum to specify how to rotate the array.
  dst: output image of the same size and depth as src image. It is an optional parameter.
'''
Rotated=cv2.rotate(img, cv2.ROTATE_90_COUNTERCLOCKWISE)
```

#Display the Rotated image

```python
cv2_imshow(Rotated)
```
