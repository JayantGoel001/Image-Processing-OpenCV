#Import Modules

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

#Load Image 

```python
#image is loaded using cv2.imread() method,here flag is 0 ,specifies to load image in GRAYSCALE mode.
'''
Syntax:
  cv2.imread(path,flag)
Parameters:
  path: string representing the path of the image to be read.
  flag: specifies the way in which image should be read.
'''
img=cv2.imread("input.png",0)
cv2_imshow(img)
```

# Set the amount of zoom

```python
# Calculate the width and height of image after zooming after applying x amount of zoom.
x=1.5
width =int(img.shape[1]*x)
height=int(img.shape[0]*x)
dim=(width,height)
```

#Apply zooming Operation




```python
# To perform zooming operation,cv2.resize() method is used.
'''
Syntax:
  cv2.resize(image,dim,interpolation)
Parameters:
  image: input image.
  dim:(width,height), determining the size of output image.
  interpolation: interpolation method to be used.
'''
zoomed=cv2.resize(img,dim,cv2.INTER_CUBIC)
```

#Display the zoomed image

```python
cv2_imshow(zoomed)
```
