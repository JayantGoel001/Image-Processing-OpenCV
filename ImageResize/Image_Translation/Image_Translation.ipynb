#Import Modules

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

#Load Image 

```python
#image is loaded using cv2.imread() method,here flag is 1 ,specifies to load a color image.
'''
Syntax:
  cv2.imread(path,flag)
Parameters:
  path: string representing the path of the image to be read.
  flag: specifies the way in which image should be read.
'''
img=cv2.imread("input.jpg",1)
cv2_imshow(img)
```

#Apply Translation operation


```python
# To perform translation,cv2.wrapAffine() method is used.
'''
Syntax:
  cv2.warpAffine(src, M, dsize,dst,flags,borderMode,borderValue)
Parameters:
  src-input image.
  M-transformation matrix.
  dsize-the size of the output image.
  flags-combination of interpolation methods
  borderMode-border pixel mode
  borderValue- border fill value; by default it is 0, beacause border fill is black by default.
'''
height, width = img.shape[:2] 
quarter_height, quarter_width = height / 4, width / 4 
T = np.float32([[1, 0, quarter_width], [0, 1, quarter_height]])
Translated = cv2.warpAffine(img, T, (width, height))
```

#Display the Translated image

```python
cv2_imshow(Translated)
```
