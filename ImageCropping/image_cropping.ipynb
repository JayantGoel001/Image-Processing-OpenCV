# Image Cropping 

```python
import cv2
from matplotlib import pyplot as plot
import numpy as nm
```

```python
img =cv2.imread("house.jfif")
plot.imshow(img)
plot.title("Original Image")
```

### Cropping across width from 30 to 150 px

```python
width_crop_1=img[:,30:150]
plot.imshow(width_crop_1)
plot.title("cropping across width from 30 to 150 px")
```

### Cropping across Height from 10 to 100 px

```python
height_crop_1=img[10:100,:]
plot.imshow(height_crop_1)
plot.title("cropping across height from 10 to 100 px")
```

### Cropping across Height 20 to 120 px and Width 90 to 210 px

```python
width_height_crop_1=img[20:120,90:210]
plot.imshow(width_height_crop_1)
plot.title("Cropping across Height 20 to 120 px and Width 90 to 210 px")
```
