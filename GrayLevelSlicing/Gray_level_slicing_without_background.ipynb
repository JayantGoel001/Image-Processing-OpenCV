## **Gray level slicing without background**

```python
import cv2
from google.colab.patches import cv2_imshow
image = cv2.imread("sloth.jpeg")
cv2_imshow(image)

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

print(gray.shape)
h,w= gray.shape

print(image[0][0])
print(gray[0][0])

for i in range(h):
    for j in range(w):
        if gray[i][j] > 50 and gray[i][j] <150:
            gray[i][j] = 255
        else:
            gray[i][j] = 0


cv2_imshow(gray)
cv2.waitKey(0)
```
