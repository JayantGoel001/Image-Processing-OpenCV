#Import Modules

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

#Load Image 

```python
#image is loaded using cv2.imread() method,here flag is 0 ,specifies to load image in GRAYSCALE mode.
'''
Syntax:
  cv2.imread(path,flag)
Parameters:
  path: string representing the path of the image to be read.
  flag: specifies the way in which image should be read.
'''
img=cv2.imread("input.png",0)
cv2_imshow(img)
```

#Apply flipping Operation




```python
# To perform flipping operation,cv2.flip() method is used.
'''
Syntax:
  cv2.flip(image,flipCode)
Parameters:
  image: input image.
  flipCode: used to specify the way, we want to flip the image.
      ( 0 : vertical flip
        1 : horizontal flip
       -1 : vertical and horizontal flip together)
'''
vertically_flipped = cv2.flip(img,flipCode=0)
horizontally_flipped = cv2.flip(img,flipCode=1)
vertically_and_horizontally_flipped = cv2.flip(img,flipCode=-1)
```

#Display the flipped image

```python
cv2_imshow(vertically_flipped)
cv2_imshow(horizontally_flipped)
cv2_imshow(vertically_and_horizontally_flipped)
```
