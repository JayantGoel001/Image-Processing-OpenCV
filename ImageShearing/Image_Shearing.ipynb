#Import Modules

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

#Load Image 

```python
#image is loaded using cv2.imread() method,here flag is 0 ,specifies to load image in GRAYSCALE mode.
'''
Syntax:
  cv2.imread(path,flag)
Parameters:
  path: string representing the path of the image to be read.
  flag: specifies the way in which image should be read.
'''
img=cv2.imread("input.png",0)
cv2_imshow(img)
```

# Set the transformation matrix and axis for shearing 

```python
# get the shape of image
rows, cols =img.shape

#shearing applied to x-axis.
M = np.float32([[1, 0.5, 0],
             	[0, 1  , 0],
            	[0, 0  , 1]])

#shearing applied to y-axis
#M = np.float32([[1,   0, 0],
#            	  [0.5, 1, 0],
#            	  [0,   0, 1]])
```

#Apply shearing Operation




```python
# To perform shearing operation,cv2.warpPerspective() method is used.
'''
Syntax:
  cv2.warpPerspective(image, dst, dsize)
Parameters:
  image: input image.
  dst: output image that has the size dsize and the same type as input image.
  dsize: size of output image
'''
sheared_img = cv2.warpPerspective(img,M,(int(cols*1.5),int(rows*1.5)))
```

#Display the sheared image

```python
cv2_imshow(sheared_img)
```
