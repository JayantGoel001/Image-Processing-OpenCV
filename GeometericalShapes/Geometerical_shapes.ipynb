# **Drawing lines in OpenCV**

```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
original_image = cv2.imread("shape.jpeg")
cv2_imshow(original_image)
```

```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
image = cv2.imread("shape.jpeg")

green = (0, 255, 0)
cv2.line(image, (0,0), (1000, 1000), green)
cv2_imshow(image)
cv2.waitKey(0)
```

# **Drawing rectangles in OpenCV**

```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow

image = cv2.imread("shape.jpeg")
green = (0, 255, 0)
cv2.rectangle(image, (80,80), (500,500), green)
cv2_imshow(image)
```

# **Drawing circles in OpenCV**

```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow

image = cv2.imread("shape.jpeg")
red = (0, 0, 255)
(centerX, centerY) = (image.shape[1]//2, image.shape[0]//2)
cv2.circle(image, (centerX, centerY), 200, red)
cv2_imshow(image)
cv2.waitKey(0)
```

# **Drawing concentric circles in openCV**

```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow

image = cv2.imread("shape.jpeg")
white = (0, 0, 255)
(centerX, centerY) = (image.shape[1]//2, image.shape[0]//2)
for r in range(0, 300, 40):
  cv2.circle(image, (centerX, centerY), r, white)
cv2_imshow(image)
cv2.waitKey(0)
```

# **Drawing random colorful circles in openCV**

```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow

image = cv2.imread("shape.jpeg")
for i in range(0, 25):
  radius = np.random.randint(5, high = 200)
  color = np.random.randint(0, high = 256, size = (3,))
  pt = np.random.randint(0, high = 300, size = (2,))
  
  color = ( int (color [ 0 ]), int (color [ 1 ]), int (color [ 2 ])) 
  cv2.circle(image, tuple(pt), radius, tuple(color), -1)
cv2_imshow(image)
cv2.waitKey(0)
```
