#Import Modules

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

#Load Image 

```python
#image is loaded using cv2.imread() method,here flag is 0 ,specifies to load image in GRAYSCALE mode.
'''
Syntax:
  cv2.imread(path,flag)
Parameters:
  path: string representing the path of the image to be read.
  flag: specifies the way in which image should be read.
'''
img=cv2.imread("input.jpg",0)
cv2_imshow(img)
```

#Apply Dilation



```python
# To perform Dilation,cv2.dilate() method is used.
'''
Syntax:
  cv2.dilate( src, kernel, iterations )
Parameters:
  src: Input image that is to be dilated.
  kernel: the matrix with which input image is convolved
  iterations: It determines how much you want to dilate the given image.
'''
kernel=np.ones((5,5),np.uint8)
Dilated=cv2.dilate(img,kernel,iterations=1)
```

#Display the Dilated image

```python
cv2_imshow(Dilated)
```
