#Import Modules

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

#Load Image 

```python
#image is loaded using cv2.imread() method,here flag is 0 ,specifies to load image in GRAYSCALE mode.
'''
Syntax:
  cv2.imread(path,flag)
Parameters:
  path: string representing the path of the image to be read.
  flag: specifies the way in which image should be read.
'''
img=cv2.imread("input.jpg",0)
cv2_imshow(img)
```

#Apply Erosion



```python
# To perform Erosion,cv2.erode() method is used.
'''
Syntax:
  cv2.erode( src, kernel, iterations )
Parameters:
  src: Input image that is to be eroded.
  kernel: the matrix with which input image is convolved
  iterations: It determines how much you want to erode the given image.
'''
kernel=np.ones((5,5),np.uint8)
Eroded=cv2.erode(img,kernel,iterations=1)
```

#Display the Eroded image

```python
cv2_imshow(Eroded)
```
