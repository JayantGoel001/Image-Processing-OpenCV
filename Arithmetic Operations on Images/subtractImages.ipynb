```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
image1 = cv2.imread(r'Images/circle1.jpg')
plt.imshow(image1)
plt.show()
```

```python
image2 = cv2.imread(r'Images/pattern1.jpg')
plt.imshow(image2)
plt.show()
```

```python
subImage = cv2.subtract(image1,image2)
```

```python
cv2.imshow('subtracted image',subImage)
plt.imshow(subImage)
plt.show()
```
