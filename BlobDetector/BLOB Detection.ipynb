```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Read image
img = cv2.imread('sunflower.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(img, cmap='gray')
```

```python
# Set up the blob detector.
blob_detector = cv2.SimpleBlobDetector_create()
```

```python
# Detect blobs from the image.
keypoints = blob_detector.detect(img)
```

```python
# Draw detected blobs as red circles.
# cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS - This method draws detected blobs as red circles and ensures that the 
# size of the circle corresponds to the size of the blob.
img_with_keypoints = cv2.drawKeypoints(img, keypoints, np.array([]), (0,0,255), cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
```

```python
# Show keypoints
plt.imshow(img_with_keypoints)
```
