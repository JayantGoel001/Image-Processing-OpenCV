# Averaging Blur

```python
import cv2
from matplotlib import pyplot as plot
import numpy as nm
```

```python
img =cv2.imread("butterfly.jfif")
gray_image=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plot.imshow(img)
plot.title("Original Image")
```

```python
averaging_blur_img=cv2.blur(gray_image,(6,6))
```

```python
plot.imshow(averaging_blur_img)
plot.title("Sobel Image X-Direction")
```
