```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import glob
```

# Let's see how to apply Fourier Transform

## Input Image

```python
# Opening image
image = cv2.imread("input/a_1.png", 0)

plt.imshow(image, cmap="gray")
```

## Applying Fourier Transform and see the Output

```python
# Calculating Fourier transform by using numpy
fft = np.fft.fft2(image)

# Magnitude Spectrum to display above fft graphically
magnitude_spectrum = 20*np.log(np.abs(fft))
# Displaying image by using OpenCV
magnitude_spectrum = np.asarray(magnitude_spectrum, dtype=np.uint8)

plt.imshow(magnitude_spectrum, cmap='gray')
```

# Appling Fourier Transform on Multiple images

```python
# Opening all the images within the folder
images = glob.iglob("input/*")
```

```python
fig = plt.figure(figsize=(10, 10))
rows = 3
columns = 3

for no, image_name in enumerate(images):
    img = cv2.imread(image_name, cv2.IMREAD_GRAYSCALE)
    
    # Calculating Fourier transform by using numpy
    fft = np.fft.fft2(img)
    # Magnitude Spectrum to display above fft graphically
    magnitude_spectrum = 20*np.log(np.abs(fft))
    # Displaying image by using OpenCV
    magnitude_spectrum = np.asarray(magnitude_spectrum, dtype=np.uint8)
    img_and_magnitude = np.concatenate((img, magnitude_spectrum), axis=1)
    
    fig.add_subplot(rows, columns, no+1)
  
    # showing image
    plt.imshow(img_and_magnitude, cmap='gray')
    plt.axis('off')
    plt.title(image_name)
```

You can observe a pattern here 
- Images of same Alphabetical character have similar fft result
- and this result can be used in image recognition
