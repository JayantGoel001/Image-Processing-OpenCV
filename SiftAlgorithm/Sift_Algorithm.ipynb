# Import Modules

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Load Image 

```python
#image is loaded using cv2.imread() method,here flag is 0 ,specifies to load image in GRAYSCALE mode.
'''
Syntax:
  cv2.imread(path,flag)
Parameters:
  path: string representing the path of the image to be read.
  flag: specifies the way in which image should be read.
'''
img=cv2.imread("charlie.jpg",0)
plt.imshow(cv2.cvtColor(img,cv2.COLOR_GRAY2RGB))
```

# Apply Sift Algorithm

```python
#1. Create an object of sift:
sift=cv2.xfeatures2d.SIFT_create()

#2. Detection and computation of Keypoints:
keypoints, descriptors =sift.detectAndCompute(img,None)

#3. The keys shown and described keypoints:
Sifted=cv2.drawKeypoints(img, keypoints, None)
```

# Display Sifted Image

```python
plt.imshow(Sifted)
```
