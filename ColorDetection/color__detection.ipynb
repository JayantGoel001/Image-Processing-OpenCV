# Import Modules

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

# Load the Image and Display it.

```python
# The image is loaded using cv2.imread() method and displayed using cv2_imshow().
'''syntax:
   cv2.imread(path, flag)
   Parameters:
   path- string representing path of the image to be read.
   Flag- specifies how the image should be read.'''
# We set flag to 1 to indicate that we want to load the image in color.

img=cv2.imread("flower.jfif",1)
cv2_imshow(img)
```

# Preprocessing The Image

```python
# The image is First converted from BGR to HSV color-space.
# This is because openCV stores color in BGR format.But,HSV color space is most suitable for color based image segmentation.
'''Syntax:
    cv2.cvtColor(src,code[,dst[,dstCn]])
    Parameters:
    src- image whose color-space is to be changed.
    code- color-space conversion code.
    dst- output image of same size and depth as that of input src image.
    dstCn- Number of channels in the destination image.
    '''
# dst and dstCn are optional Parameters.So,we won't use them here.

final_img=cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
```

# Create Mask for color detection

```python
# Use color Picker to define the upper and lower range of color need to be detected.
'''Choose hue value for a particular color as x degree and define the HSV range as
   lower_hsv=((x/2)-5,100,100) 
   upper_hsv=((x/2)+5,255,255) and create numpy array for them.'''
# Here,We want to detect all the green colored regions,for which hue value is 75 degrees.

lower=np.array([32,100,100])
upper=np.array([42,255,255])

# Use cv2.inRange() function to create mask over area of interest.

area_of_interest=cv2.inRange(final_img,lower,upper)
cv2_imshow(area_of_interest)
```

# Apply mask and Display the result 


```python
# Bitwise operations is used for image masking.
'''Syntax:
   cv2.bitwise_and(src1,src2,destination,mask)
   Parameters:
   src1- first input image array
   src2- second input image array
   destination- Output array similar in dimension and type to that of input 
   mask- operation mask/area of interest'''

result=cv2.bitwise_and(img,img,mask=area_of_interest)
cv2_imshow(result)
```
